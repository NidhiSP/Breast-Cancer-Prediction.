### **Some Risk Factors for Breast Cancer**

The following are some of the known risk factors for breast cancer. However, most cases of breast cancer cannot be linked to a specific cause. Talk to your doctor about your specific risk.

**Age**

The chance of getting breast cancer increases as women age. Nearly 80 percent of breast cancers are found in women over the age of 50.

**Personal history of breast cancer**

A woman who has had breast cancer in one breast is at an increased risk of developing cancer in her other breast.

**Family history of breast cancer**

A woman has a higher risk of breast cancer if her mother, sister or daughter had breast cancer, especially at a young age (before 40). Having other relatives with breast cancer may also raise the risk.

**Genetic factors**

Women with certain genetic mutations, including changes to the BRCA1 and BRCA2 genes, are at higher risk of developing breast cancer during their lifetime. Other gene changes may raise breast cancer risk as well.

**Childbearing and menstrual history**
The older a woman is when she has her first child, the greater her risk of breast cancer. Also at higher risk are:
*   Women who menstruate for the first time at an early age (before 12)
*   Women whoâ€™ve never had children
*   Women who go through menopause late (after age 55)

```python
### Divide data into features and target
x = data.iloc[:,3:]   ## features
y = data.iloc[:,2]    ## target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80,random_state=1)
```

```python
import tensorflow.keras as tf
```

```python
### build the model (Create the object for model)
model = tf.models.Sequential()  ## blank model
## Adding the layers
model.add(tf.layers.Dense(16,input_dim=14,activation="relu"))  ## two layers - input layer,hidden
model.add(tf.layers.Dense(32,activation='relu'))  ## 2nd hidden layer
model.add(tf.layers.Dense(1,activation='sigmoid'))   ## output layer
## Compile the model
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
### Train the model
model.fit(xtrain,ytrain,epochs=500)
```

```python
### Test/Prediction
ypred = model.predict(xtest)
ypred = ypred.round()
ypred
```

```python
df = pd.DataFrame()
df['Ytest'] = ytest
df['Ypred'] = ypred
df
```

```python
model.evaluate(xtest,ytest)
```

```python
### Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(ytest,ypred)
a
```
